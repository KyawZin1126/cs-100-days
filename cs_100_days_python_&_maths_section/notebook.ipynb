## Limit

```python
from sympy import symbols,factor

x = symbols('x')
expression = x ** 2 - 25
factor_expression = factor(expression)
print(factor_expression)
```

```python
x = symbols('x')
expression = x ** 2 + 5 * x + 6
factor_expression = factor(expression)
print(factor_expression)
```

```python
x = symbols('x')
expression = 3 * x ** 2 + 6 * x
factor_expression = factor(expression)
print(factor_expression)
```
