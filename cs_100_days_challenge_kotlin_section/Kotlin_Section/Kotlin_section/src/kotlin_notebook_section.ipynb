## When Statement

```kotlin
var a = 10
when(a){
    1 -> println("a is 1")
    2 -> println("a is 2")
    3,4,5 -> println("a is between 3 and 5")
    else -> {
        println("a is nothing")
        println("a is not include in here")
    }
}

//when as expression

var result = when(4){
    1 -> "One"
    2 -> "Two"
    3 -> "Three"
    4,5 -> "Four or Five"
    6 -> {
        "Six"
        600
    }
    else -> "Nothing"
}

println(result)
```
